```python
from selenium import webdriver
import pandas as pd
import time

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

```python
PATH = "/Users/aoxiangsuen/chromedriver"
driver = webdriver.Chrome(PATH)
```

```python
# initialize empty dataframe
df = pd.DataFrame(columns = ['title', 'img_url', 'price', 'colour', 'description'])

url_list = ['https://www.asos.com/men/jackets-coats/cat/?cid=3606&nlid=mw|clothing|shop%20by%20product&page=',
            'https://www.asos.com/men/jeans/cat/?cid=4208&nlid=mw|clothing|shop%20by%20product&page=',
            'https://www.asos.com/men/tracksuits/joggers/cat/?cid=14274&nlid=mw|clothing|shop%20by%20product&page=',
            'https://www.asos.com/men/jumpers-cardigans/cat/?cid=7617&nlid=mw|clothing|shop%20by%20product&page=',
            'https://www.asos.com/men/loungewear/cat/?cid=18797&nlid=mw|clothing|shop%20by%20product&page=',
            'https://www.asos.com/men/polo-shirts/cat/?cid=4616&nlid=mw|clothing|shop%20by%20product&page=',
            'https://www.asos.com/men/shirts/cat/?cid=3602&nlid=mw|clothing|shop%20by%20product&page=', 
            'https://www.asos.com/men/shorts/cat/?cid=7078&nlid=mw|clothing|shop%20by%20product&page=',
            'https://www.asos.com/men/suits/cat/?cid=5678&nlid=mw|clothing|shop%20by%20product&page=',
            'https://www.asos.com/men/t-shirts-vests/cat/?cid=7616&nlid=mw|clothing|shop%20by%20product&page=',
            'https://www.asos.com/men/tracksuits/cat/?cid=26776&nlid=mw|clothing|shop%20by%20product&page=',
            'https://www.asos.com/men/trousers-chinos/cat/?cid=4910&nlid=mw|clothing|shop%20by%20product&page=']


for url in url_list:
    
    for page in range(30):
        
        page_num = str(page + 1)
    
        driver.delete_all_cookies()
        driver.get(url+page_num)

        # Get product IDs
        ids = driver.find_elements_by_xpath("//a[@class='_3TqU78D']")

        for i in range(len(ids)):

            driver.delete_all_cookies()
    
            ids_reload = driver.find_elements_by_xpath("//a[@class='_3TqU78D']")
            
            try:
                ids_reload[i].click()
            except:
                pass
            
            driver.implicitly_wait(3) 

            # get info
            
            try:
                img_url = driver.find_elements_by_xpath("//*[@class='img']")[1].get_attribute('src')
            except:
                driver.get(url+page_num)
                try:
                    img_url = driver.find_elements_by_xpath("//*[@class='img']")[1].get_attribute('src')
                except:
                    try:
                        img_url = driver.find_element_by_xpath('//*[@class="img"]').get_attribute('src')
                    except:
                        img_url = 0
                        pass
                
            try:
                title = driver.find_element_by_xpath('//*[@id="aside-content"]/div[1]/h1').text
            except:
                title = 0
                pass
                
            try:
                price = driver.find_element_by_xpath('//*[@id="product-price"]/div/span[2]/span[4]/span[1]').text
            except:
                price = 0
                pass

            try:
                colour = driver.find_element_by_xpath('//*[@id="product-colour"]/section/div/div/span').text
            except:
                colour = 0
                pass

            try:
                description = driver.find_element_by_xpath('//*[@id="product-details-container"]/div[2]/div[2]/p').text
            except:
                driver.get(url+page_num)
                try:
                    description = driver.find_element_by_xpath('//*[@id="product-details-container"]/div[2]/div[2]/p').text
                except:
                    description = 0
                    pass
                
            dic = {'title':title, 'img_url':img_url, 'price':price , 'colour':colour, 'description':description}
            series = pd.Series(dic)

            df = df.append(series, ignore_index=True)

            # back up to csv file
            if len(df) % 10 == 0:
                df.to_csv('asos_men.csv')

            else:
                pass

            driver.get(url+page_num)

            print("Page {} Index {} Total {}: {}".format(page_num, i+1, df.index.stop, title))
        
        if len(ids) != 72:
            break
        
        else:
            pass

    print(print("="*20,"next url","="*20))
```
