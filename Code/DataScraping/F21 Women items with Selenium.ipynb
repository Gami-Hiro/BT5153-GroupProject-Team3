```python
# Import packages
import selenium
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
#Install Driver
driver = webdriver.Chrome(ChromeDriverManager().install())
```

# Forever 21 Women Items

## 1) Retrieve all product links

```python
# Specify Search URL 
search_url = "https://www.forever21.com/us/shop/catalog/category/f21/app-main?cgid=app_main&start={}&sz=120" 
```

```python
# extract for all 55 pages
# each page displays 120 items (55 * 120 = 6600 items)
page_ind = list(range(0, 6600, 120))
print(page_ind)
print(len(page_ind))
```

```python
f21_url = "https://www.forever21.com/us/shop/catalog/category/f21/app-main?cgid=app_main&start={}&sz=120" 
links = []

for page in page_ind:
    
    driver.get(search_url.format(page))
    find_links = driver.find_elements_by_xpath("//a[@data-product-swatch]")
    
    for i in find_links:
        url = i.get_attribute("href")
        links.append(url)
        print(url)
```

```python
# check that product links are unique
print('Number of links = {}'.format(len(links)))
print('Number of unique links = {}'.format(len(set(links))))
```

```python
# select only unique product links
links = list(set(links))
print(len(links))
links
```

## 2) Loop through all product links to scrape required information

### b) Putting it all together in a loop

```python
# define dataframe
f21_women = pd.DataFrame(columns=['Title', 'Price', 'Gender', 'Category', 'Colour', 'Description', 'Material', 'First_Image_URL'])
f21_women
```

```python
# loop through links
i = 0

for link in links:
    
    # initialize fields
    title = 'NA'
    price = 'NA'
    category = 'NA'
    colour = 'NA'
    desc = 'NA'
    material = 'NA'
    img = 'NA'
    
    # access link in driver
    driver.get(link)
    
    # get product title        
    find_title = driver.find_elements_by_xpath("//h1[contains(@data-product-component, 'name')]")
    if len(find_title) > 0:
        title = find_title[0].text
    
    # get product price (price before discount)      
    price_path = "//div[@class = 'pdp-main__price body-type']//span[contains(@itemprop, 'price') and not(contains(@class, 'discount'))]"
    find_price = driver.find_elements_by_xpath(price_path)
    if len(find_price) > 0:
        price = find_price[0].text

    # get product category
    cat_path = "//div[@class='pdp__breadcrumbs display--small-only']//li[last()-1]/a[@title]"
    find_category = driver.find_elements_by_xpath(cat_path)
    if len(find_category) > 0:
        category = find_category[0].text
        if category == '':
            category = find_category[0].get_attribute('title')
    
    # get product colour
    colour_path = "//div[@class='product-attribute__selected-value form-control-label ']/span"
    find_colour = driver.find_elements_by_xpath(colour_path)
    if len(find_colour) > 0:
        colour = find_colour[0].text
    
    # get product description
    desc_path = "//div[@class='pdp__details']//section[h3='Details']/div[@class='d_content']"
    find_desc = driver.find_elements_by_xpath(desc_path)
    if len(find_desc) > 0:
        desc = ''.join([i.text for i in find_desc])
    
    # get product material   
    material_path = "//div[@class='pdp__details']//section[h3='Content + Care']/div[@class='d_content']/p[position()<last()]"
    find_material = driver.find_elements_by_xpath(material_path)
    if len(find_material) > 0:
        material = ''.join([i.text for i in find_material])

    # get image URL   
    img_path = "//li[@class='product-gallery__item slick-slide slick-current slick-active']//button[@tabindex='0']/img"
    find_img = driver.find_elements_by_xpath(img_path)
    if len(find_img) > 0:
        img = find_img[0].get_attribute('src')
    
    # fill up dataframe
    f21_women.loc[i, 'Title'] = title
    f21_women.loc[i, 'Price'] = price
    f21_women.loc[i, 'Category'] = category
    f21_women.loc[i, 'Colour'] = colour
    f21_women.loc[i, 'Description'] = desc
    f21_women.loc[i, 'Material'] = material
    f21_women.loc[i, 'First_Image_URL'] = img
    
    # add to counter
    i += 1
    print(i)
```

## 3) Clean data

### a) Remove non-clothing categories

```python
# examine distribution of categories
f21_women['Category'].value_counts()
```

```python
# remove non clothing categories
# kept 'branded shop' and 'women' as the items are clothing items
# note: pajamas/sleep sets under 'Home' are not included
cat_excl = ['Home', 'Lingerie + Sleepwear', 'Shoes', 'Swimwear', 'Caps + Hats', 'Jewelry', 'Socks + Tights', 'Bags + Backpacks',
            'Hair Accessories', 'Sunglasses + Readers', 'Scarves + Gloves', 'Belts', 'Accessories', 'Home + Tech', 'Skincare',
            'Hats + Beanies']
f21_women_clothing = f21_women[~f21_women['Category'].isin(cat_excl)]
f21_women_clothing['Category'].value_counts()
```

```python
len(f21_women_clothing)
```

### b) Clean description & material

```python
# remove the phrase on being an independent brand in the description

# sample
f21_women_clothing[f21_women_clothing['Title'] == 'Acid Wash Boxy Denim Jacket']['Description'][9]
```

```python
# remove phrase for entire dataset
phrase1 = '\n- This is an independent brand and not a Forever 21 branded item.'
phrase2 = '\(This is an independent brand and not a Forever 21 branded item.\)'
f21_women_clothing['Description'] = f21_women_clothing['Description'].replace(phrase1, '', regex=True)
f21_women_clothing['Description'] = f21_women_clothing['Description'].replace(phrase2, '', regex=True)
```

```python
f21_women_clothing[f21_women_clothing['Title'] == 'Cotton-Blend Cropped Cami']['Description']
```

```python
# remove style deals phrase in description for entire dataset
f21_women_clothing['Description'] = f21_women_clothing['Description'].replace('Style Deals - ', '', regex=True)
f21_women_clothing[f21_women_clothing['Title'] == 'Cotton-Blend Cropped Cami']['Description'][1053]
```

```python
# check for additional bullet points not pertaining to description
f21_women_clothing['desc_add'] = f21_women_clothing['Description'].str.split('\n-').str[1]
f21_women_clothing['desc_add'].value_counts() # looks like can be removed
```

```python
# remove additional bullet points not pertaining to description
f21_women_clothing['Description'] = f21_women_clothing['Description'].str.split('\n-').str[0]
```

```python
# drop desc add column
f21_women_clothing.drop(['desc_add'], axis=1, inplace=True)
```

```python
# remove mention of 'matching xx available' in brackets
to_replace = {
    '\(Matching top available.\)': '',
    '\(Matching bottoms available.\)': '',
    '\(Matching mask available.\)': '',
    '\(Matching bottoms available. Complete the look with 00428975.\)': '',
    '\(Matching top available. Complete the look with 00428970.\)': '',
    '\(Matching cardigan and shorts available.\)': '',
    '\(Matching top & cardigan available.\)': '',
    '\(PLEASE NOTE: This is NOT a medical-grade mask and is not intended as a replacement for medical-grade equipment or other recommended measures to stop the community spread of any viruses, including social distancing, washing your hands and refraining from touching your face.\)': '',
    '< p>- This is an independent brand and not a Forever 21 branded item.': ''
}

f21_women_clothing['Description'] = f21_women_clothing['Description'].replace(to_replace, regex=True)
```

```python
# remove the phrase on machine/hand wash from material
f21_women_clothing['Material'].fillna("NA", inplace = True) # need to replace NA with string first
f21_women_clothing[f21_women_clothing['Material'].str.contains("wash")]
```

```python
# remove washing phrase in material for entire dataset
phrase1 = '- Machine wash cold'
phrase2 = '- Hand wash cold'
f21_women_clothing['Material'] = f21_women_clothing['Material'].replace(phrase1, '', regex=True)
f21_women_clothing['Material'] = f21_women_clothing['Material'].replace(phrase2, '', regex=True)
```
